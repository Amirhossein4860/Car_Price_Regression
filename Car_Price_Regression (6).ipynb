# Handling Non Numeric Features

The LabelEncoder class from the sklearn.preprocessing module is used to convert categorical data into numerical form. It is particularly helpful when machine learning algorithms require input features to be numerical, as most models cannot process raw categorical data directly.

---

**Purpose:**

- Convert non-numerical labels (categories) into integers.

- Ensure consistent mapping of categorical values to numerical representations.

---

**How it Works:**

1. Each unique category in a column is assigned a unique integer value.

2. This mapping is stored internally in the LabelEncoder object, ensuring consistency during transformations.

3. The resulting numerical values can then be used as input for machine learning algorithms.

---

**Key Functions:**

- fit: Identifies the unique categories in the data and creates a mapping from category to integer.

- transform: Converts the original categorical data into numerical labels using the created mapping.

- fit_transform: Combines the fit and transform steps into one for convenience.

- inverse_transform: Converts the numerical labels back into their original categorical values.

---

**When to Use:**

- When the categorical data has no intrinsic order, and a simple numerical  representation is sufficient.

- When you want a lightweight encoding for nominal features.



```python
from sklearn.preprocessing import LabelEncoder
```

This line initializes a LabelEncoder object, which will be used to transform categorical data into numerical format.

```python
encoder = LabelEncoder()
```

This block of code applies label encoding to multiple categorical columns in the DataFrame df. Each column is transformed into numerical format, where each unique category is assigned a unique integer value.

**Explanation:**

- encoder.fit_transform(column):

 - Fits the LabelEncoder to the specific column, learning the unique categories.

 - Transforms the categories into integers based on the learned mapping.

**Purpose:**

- Converts categorical data into numerical format, making it suitable for machine learning algorithms.

- Ensures consistent encoding for each column.

**Columns Transformed:**

- Manufacturer, Model, Category, Leather interior, Fuel type, Gear box type, Drive wheels, Doors, Wheel, Color, Engine volume

 - Each of these categorical columns is encoded into integers.

---


**Why This is Done:**

- Machine learning models generally require numerical input; categorical data cannot be processed directly.

- Label encoding helps to prepare categorical data for use in predictive modeling.

```python
df['Manufacturer'] = encoder.fit_transform(df['Manufacturer'])
df['Model'] = encoder.fit_transform(df['Model'])
df['Category'] = encoder.fit_transform(df['Category'])
df['Leather interior'] = encoder.fit_transform(df['Leather interior'])
df['Fuel type'] = encoder.fit_transform(df['Fuel type'])
df['Gear box type'] = encoder.fit_transform(df['Gear box type'])
df['Drive wheels'] = encoder.fit_transform(df['Drive wheels'])
df['Doors'] = encoder.fit_transform(df['Doors'])
df['Wheel'] = encoder.fit_transform(df['Wheel'])
df['Color'] = encoder.fit_transform(df['Color'])
df['Engine volume'] = encoder.fit_transform(df['Engine volume'])
```

# **Box Plot**

A box plot (or box-and-whisker plot) is a visualization used to display the distribution, spread, and skewness of data through quartiles. It highlights the median, interquartile range (IQR), and potential outliers for a variable.

**Median:** The central line in the box represents the median (50th percentile) of the data.

**IQR (Interquartile Range):** The box represents the middle 50% of the data, spanning from the first quartile (25th percentile) to the third quartile (75th percentile).

**Whiskers:** The lines (whiskers) extending from the box show the range within 1.5 times the IQR.

**Outliers:** Data points outside this range are plotted as individual points and considered potential outliers.

```python
n_cols = 2
n_rows = (len(numeric_columns) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
axes = axes.flatten()

for i, column in enumerate(numeric_columns):
    sns.boxplot(x=df[column], ax=axes[i])
    axes[i].set_title(f'Box Plot of {column}')

if len(numeric_columns) < len(axes):
    for j in range(len(numeric_columns), len(axes)):
        axes[j].set_visible(False)

plt.tight_layout()
plt.savefig('numeric_columns_boxplots.png')
plt.show()
```

# **Correlation Matrix**

df.corr() computes the correlation matrix for all numerical features in the DataFrame df. Each value in this matrix represents the correlation coefficient between two features, indicating how strongly they are linearly related (with values ranging from -1 to 1).

```python
corr = df.corr()
```

np.ones_like(corr, dtype=bool) creates a matrix of the same shape as corr, filled with True values.

np.triu(...) then converts this matrix into an "upper triangular" form, keeping only the upper triangle of the matrix as True and setting all values below the diagonal to False.

This mask is often used in visualization (e.g., with a heatmap) to hide the lower triangle of the correlation matrix, as it’s typically symmetrical across the diagonal.

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
```

**Explanation of the Heatmap**

A correlation heatmap is a graphical representation of the correlation matrix, where each cell in the matrix displays the correlation coefficient between two variables. In this heatmap, colors represent the strength and direction of the correlation:

Red shades indicate positive correlations (as one variable increases, the other tends to increase as well).

Blue shades indicate negative correlations (as one variable increases, the other tends to decrease).

Intensity of color shows the strength of the correlation: darker colors imply stronger correlations, whether positive or negative.

Cells are masked to show only the upper triangle, given the matrix is symmetric around the diagonal.

```python
f, ax = plt.subplots(figsize=(23, 19))
cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap='coolwarm', vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5} , annot=True)

plt.show()
```

---

**Key Concept:**

- Polynomial Feature Generation:

 - This technique generates new features by considering powers of the original features. For example, if you have features x1 and x2 , PolynomialFeatures can create new features like x1,x2 , and interaction terms such as x1,x2.

- Interaction Features:

 - Interaction terms are combinations of the input features that capture relationships between them. These features help the model learn more complex patterns, especially in non-linear datasets.

- Degree of Polynomial:

 - The degree of the polynomial defines how many levels of powers of the features are generated. For instance, a degree of 2 generates the original features, squared features, and interaction terms, while a degree of 3 would also add cubic terms and further interactions.

---


**How It Works:**

1. Feature Expansion:

 - PolynomialFeatures transforms the input features into higher-degree polynomial terms. This allows a linear model (like linear regression) to capture non-linear relationships by creating polynomial interactions.

2. Fit and Transform:

 - When applied, PolynomialFeatures learns the degree of transformation from the data and applies it to create the expanded feature set.

3. Use Case with Regression Models:

 - For example, in a linear regression model, adding polynomial features allows the model to fit a curve instead of a straight line, thereby improving performance on datasets with non-linear trends.

---


**Advantages:**

- Capturing Non-Linear Relationships: It allows linear models to fit non-linear data by expanding the feature space to include polynomial terms.

- Improved Model Flexibility: By adding polynomial features, the model can learn more complex patterns, improving its ability to capture non-linear relationships in the data.

---

**When to Use:**

- Non-Linear Data: When you suspect that the relationship between the features and the target variable is not linear.

- Feature Engineering: When you need to enhance a linear model's capacity to capture more complex relationships without switching to non-linear models.

# Lasso

**Source:** sklearn.linear_model

---

**Purpose:**

Lasso (Least Absolute Shrinkage and Selection Operator) is a linear regression model that uses L1 regularization to improve model performance by penalizing the absolute values of the coefficients. It is particularly useful for feature selection and reducing overfitting.

---

**Key Concept:**

- L1 Regularization:

 - Adds a penalty equal to the absolute value of the coefficients to the loss function.

 - Encourages sparsity by shrinking less important feature coefficients to exactly zero, effectively selecting only the most relevant features.

- Feature Selection:

 - By shrinking some coefficients to zero, Lasso inherently performs feature selection, identifying the most significant predictors while discarding irrelevant ones.

- Trade-Off Control (Alpha):

 - The alpha parameter controls the strength of the regularization.

   - Higher alpha values increase regularization, leading to more coefficients being shrunk to zero.

   - Lower alpha values reduce the regularization effect, making the model behave more like standard linear regression.

---

**How It Works:**

1. Loss Function with L1 Regularization:

 - The regularization term (𝛼 ∑ ∣ Coefficients ∣ ) penalizes large coefficients, helping prevent overfitting.

2. Coefficient Shrinkage:

 - During training, Lasso reduces the coefficients of less important features to zero, effectively removing them from the model.

3. Sparse Model:

 - The resulting model is simpler, with fewer active features, which improves interpretability and reduces overfitting.

---

**Advantages:**

- Feature Selection: Automatically removes irrelevant or redundant features, making the model more interpretable.

- Handles High-Dimensional Data: Effective for datasets with many features, even when the number of features exceeds the number of samples.

- Reduces Overfitting: By shrinking coefficients, it prevents the model from fitting noise in the data.

---

**When to Use:**

- High-Dimensional Data: When the dataset has many features, and you want to select only the most important ones.

- Preventing Overfitting: When the model is prone to overfitting due to irrelevant or highly correlated features.

- Interpretability: When a simpler model with fewer features is desired.

---

**Summary:**

- Lasso is a linear regression model with L1 regularization that performs feature selection by shrinking coefficients of less important features to zero. It is ideal for high-dimensional datasets, helps reduce overfitting, and improves model interpretability by simplifying the feature set. The strength of regularization is controlled by the alpha parameter, allowing flexibility in balancing model complexity and accuracy.

# Ridge

**Source:** sklearn.linear_model

---

**Purpose:**

Ridge is a linear regression model that incorporates **L2 regularization** to improve model performance by penalizing the square of the coefficients. It helps reduce overfitting and improves the generalizability of the model by shrinking large coefficients.

---

**Key Concept:**

- **L2 Regularization:**
  - Adds a penalty equal to the square of the magnitude of coefficients to the loss function.
  - Encourages the model to keep all coefficients small but does not shrink them to zero, unlike Lasso.

- **Trade-Off Control (Alpha):**
  - The **alpha** parameter controls the strength of the regularization.
    - Larger values of alpha increase regularization, leading to smaller coefficients and simpler models.
    - Smaller values of alpha reduce regularization, making the model closer to standard linear regression.

- **Bias-Variance Trade-Off:**
  - Ridge regression reduces variance in the model at the cost of a slight increase in bias, leading to better performance on unseen data.

---

**How It Works:**

1. **Loss Function with L2 Regularization:**
   - Ridge minimizes the following:
     \[
     \text{Loss} = \text{Residual Sum of Squares} + \alpha \sum \text{Coefficients}^2
     \]
   - The regularization term (\(\alpha \sum \text{Coefficients}^2\)) penalizes large coefficients, preventing overfitting.

2. **Coefficient Shrinkage:**
   - During training, Ridge shrinks the coefficients of less important features without setting them to zero, unlike Lasso.

3. **Balanced Model Complexity:**
   - The model balances fitting the training data and generalizing to new data by controlling the complexity through regularization.

---

**Advantages:**

- **Reduces Overfitting:** By penalizing large coefficients, Ridge prevents the model from fitting noise in the training data.
- **Handles Multicollinearity:** It is effective when features are highly correlated by distributing the impact among them.
- **Retains All Features:** Unlike Lasso, Ridge does not eliminate features but instead shrinks their coefficients.

---

**When to Use:**

- **High-Dimensional Data:** When the dataset has many features, especially with multicollinearity.
- **Preventing Overfitting:** When a simple linear regression model overfits the training data.
- **Stable Coefficients:** When it is important to keep all features in the model, even if their impact is small.

---

**Summary:**

Ridge regression is a linear model with **L2 regularization** that shrinks coefficients to prevent overfitting while retaining all features. It is particularly useful for datasets with multicollinearity or when the model needs to generalize well to unseen data. The regularization strength is controlled by the alpha parameter, allowing flexibility in balancing model complexity and performance.

# SVR

**Source:** sklearn.svm

---

**Purpose:**

SVR (Support Vector Regressor) is a supervised machine learning algorithm used for regression tasks. It is based on the Support Vector Machine (SVM) algorithm and aims to find a hyperplane that best fits the data within a specified margin of tolerance.

---

**Key Concept:**

- **Support Vector Machines (SVM):**
  - SVR uses the SVM concept to predict continuous target variables by finding a hyperplane in a high-dimensional space.
  
- **Epsilon Margin:**

  - An epsilon-tube is defined around the hyperplane, and the algorithm focuses on fitting the data within this margin, ignoring deviations within the tolerance.

- **Kernel Trick:**

  - Allows SVR to handle non-linear relationships by transforming the input features into higher-dimensional spaces using kernel functions. Common kernels include:
    - Linear
    - Polynomial
    - Radial Basis Function (RBF)
    - Sigmoid

- **Regularization Parameter (C):**
  - Controls the trade-off between achieving a low error on the training data and maintaining a smooth hyperplane. Higher values of C result in a tighter fit to the training data, while lower values emphasize a smoother model.

---

**How It Works:**

1. **Margin Construction:**

   - SVR creates a hyperplane and a margin (epsilon-tube) around it. Points within the margin are ignored, focusing only on the most relevant data points (support vectors).

2. **Kernel Transformation:**
   - If the relationship between the input features and target variable is non-linear, SVR uses a kernel function to map the data into a higher-dimensional space where a linear hyperplane can be fitted.

3. **Prediction:**
   - During prediction, SVR calculates the position of a data point relative to the hyperplane and estimates the continuous target variable.

---

**Advantages:**

- **Handles Non-Linear Relationships:** Kernels enable SVR to model complex patterns in the data.

- **Robust to Outliers:** The epsilon margin allows SVR to ignore small deviations in the data.

- **Versatile:** Works well for high-dimensional datasets and supports a variety of kernel functions.

---

**When to Use:**

- **Non-Linear Data:** When the relationship between features and the target variable is non-linear.

- **Small to Medium Datasets:** SVR can be computationally expensive for very large datasets.
- **High-Dimensional Data:** Effective in cases where the number of features is large compared to the number of samples.

---

**Summary:**

SVR is a powerful regression algorithm that extends Support Vector Machines for predicting continuous target variables. Its ability to handle non-linear relationships and outliers through kernel functions and the epsilon margin makes it a versatile choice for various regression tasks.

- Performance Metric:

# Decision Tree

**Source:** sklearn.tree

---

**Purpose:**

DecisionTreeRegressor is a supervised machine learning algorithm used for regression tasks. It predicts continuous target variables by learning simple decision rules inferred from the input features. The algorithm partitions the data into smaller subsets based on feature values, ultimately forming a tree structure.

---

**Key Concept:**

- **Tree-Based Model:**
  - A decision tree splits the dataset into smaller regions by creating branches based on feature thresholds. Each leaf node represents a predicted value.

- **Recursive Partitioning:**
  - The algorithm recursively splits the data to minimize the error (e.g., Mean Squared Error) within each partition.

- **Overfitting Control:**
  - Hyperparameters such as max_depth, min_samples_split, and min_samples_leaf help control the tree's growth and reduce overfitting.

---

**How It Works:**

1. **Tree Construction:**
   - The algorithm starts with the entire dataset at the root and splits it into subsets based on the feature that minimizes the error (e.g., variance reduction).

2. **Node Splitting:**
   - At each node, the best split is determined by evaluating all possible splits for all features.

3. **Prediction:**
   - The final prediction for a new data point is based on the average target value of the training data in the leaf node where the point lands.

---

**Advantages:**

- **Interpretability:**
  - The model is easy to visualize and interpret, as decisions are based on simple if-then rules.

- **Handles Non-Linear Relationships:**
  - Decision trees can capture complex, non-linear relationships between features and the target variable.

- **No Scaling Required:**
  - The model does not require feature scaling or standardization.

- **Feature Selection:**
  - Automatically selects the most important features by splitting on the most informative ones first.

---

**When to Use:**

- **Non-Linear Data:**
  - When the relationship between the input features and target variable is non-linear or complex.

- **Feature Importance Analysis:**
  - When you want to understand which features are most relevant for predictions.

- **Small to Medium-Sized Datasets:**
  - Decision trees perform well on smaller datasets but can overfit on large datasets without proper regularization.

---

**Summary:**

DecisionTreeRegressor is a flexible and interpretable regression algorithm that works well for non-linear relationships and datasets with complex interactions. It partitions the data into smaller regions, forming a tree structure that predicts the target variable. Proper tuning of tree parameters ensures optimal performance and prevents overfitting.

# Random Forest Regressor

**Source:** sklearn.ensemble

---

**Purpose:**

- RandomForestRegressor is a supervised machine learning algorithm used for regression tasks. It is an ensemble method that combines the predictions of multiple decision trees to improve accuracy and reduce overfitting.

- By averaging the predictions of individual trees, it creates a more robust and generalized model compared to a single decision tree.

---

**Key Concepts:**

- **Ensemble Learning:**
  - Combines the outputs of multiple models (decision trees) to improve overall performance.

- **Bagging (Bootstrap Aggregating):**
  - Each tree is trained on a randomly sampled subset of the training data (with replacement). This promotes diversity among the trees and reduces variance.

- **Random Feature Selection:**
  - At each split, the algorithm considers a random subset of features to find the best split. This further decorrelates the trees and improves generalization.

- **Prediction Aggregation:**
  - For regression tasks, the model averages the predictions of all trees to produce the final output.

---

**How It Works:**

1. **Data Sampling:**
   - Random subsets of the training data are drawn (with replacement) to train individual decision trees.

2. **Tree Construction:**
   - Each tree is built using the sampled data and considers a random subset of features at each split, ensuring diversity among trees.

3. **Prediction:**
   - For regression tasks, the predictions from all trees are averaged to produce the final output.

---

**Parameters:**

- **n_estimators:**
  - The number of decision trees in the forest (default: 100). Increasing this value can improve accuracy but increases computational cost.

- **max_depth:**
  - Limits the maximum depth of each tree. Controls tree size to prevent overfitting.

- **min_samples_split and min_samples_leaf:**
  - Control the minimum number of samples required to split a node or be in a leaf node, reducing overfitting by ensuring splits are meaningful.

- **max_features:**
  - The number of features to consider when looking for the best split. Common options include:
    - 'sqrt': Square root of the total number of features.
    - 'log2': Logarithm (base 2) of the total number of features.
    - None: Considers all features.

- **bootstrap:**
  - Whether to use bootstrap sampling (default: True). If False, the entire dataset is used to train each tree.

---

**Advantages:**

- **Reduces Overfitting:**
  - By averaging the predictions of multiple trees, the model generalizes better to unseen data.

---

**When to Use:**

# K-Neighbors Regressor

**Source:** sklearn.neighbors

---

**Purpose:**

KNeighborsRegressor (KNN Regressor) is a supervised machine learning algorithm used for regression tasks. It predicts the target variable for a given input by averaging the target values of its nearest neighbors in the feature space.

---

**Key Concept:**

- **Instance-Based Learning:**
  - KNN is a lazy learning algorithm, meaning it doesn't explicitly learn a model during training. Instead, it uses the training data directly for making predictions.

- **Neighbors:**
  - The algorithm finds the **k** nearest data points (neighbors) in the training dataset based on a specified distance metric (e.g., Euclidean distance).

- **Weighted or Uniform Averaging:**
  - Predictions are made by averaging the target values of the **k** nearest neighbors. Optionally, closer neighbors can be given more weight in the averaging process.

- **Distance Metrics:**
  - KNN uses distance metrics to determine the nearest neighbors. Common metrics include:
    - Euclidean Distance
    - Manhattan Distance
    - Minkowski Distance
    - Hamming Distance (for categorical data)

---

**How It Works:**

1. **Finding Neighbors:**
   - For a given test sample, the algorithm calculates the distance to every data point in the training set and selects the **k** closest points.

2. **Prediction:**
   - The target variable for the test sample is predicted as the average of the target values of its nearest neighbors.

3. **Distance Weighting:**
   - Optionally, neighbors closer to the test sample can be assigned higher weights, influencing the prediction more than distant neighbors.

---

**Advantages:**

- **Simple and Intuitive:** KNN is easy to understand and implement, making it a popular choice for quick and straightforward regression tasks.

- **Non-Parametric:** KNN does not make any assumptions about the data distribution, allowing it to model complex, non-linear relationships.

- **Versatile:** The algorithm can handle both regression and classification tasks with appropriate adjustments.

---

**When to Use:**

- **Small to Medium Datasets:** KNN can be computationally expensive for very large datasets as it requires calculating distances for every test sample.

- **Non-Linear Data:** KNN performs well when the relationship between features and the target variable is non-linear.

- **No Explicit Training Phase Needed:** KNN is suitable when you want to defer the bulk of computations to the prediction phase.

---

**Summary:**

KNeighborsRegressor is an intuitive, non-parametric algorithm that uses the training data directly to make predictions by averaging the target values of the nearest neighbors. Its simplicity, flexibility, and ability to model non-linear relationships make it a versatile choice for various regression tasks, especially for smaller datasets where computational efficiency is less of a concern.

# **MLPRegressor**

**Source:** sklearn.neural_network

---

**Purpose:**

MLPRegressor (Multi-Layer Perceptron Regressor) is a supervised machine learning algorithm used for regression tasks. It belongs to the class of neural network models and learns complex relationships between input features and the target variable through multiple layers of neurons.

---

**Key Concept:**

- **Artificial Neural Networks (ANNs):**
  - MLPRegressor is a feedforward neural network that consists of:
    - **Input Layer:** Accepts the input features.
    - **Hidden Layers:** Processes the input through weighted connections and non-linear activation functions to capture complex patterns.
    - **Output Layer:** Produces the continuous target variable predictions.

- **Backpropagation:**
  - The model uses backpropagation to minimize the loss function by adjusting weights and biases through gradient descent.

- **Activation Functions:**
  - Activation functions introduce non-linearity into the model, enabling it to learn complex patterns. Common activation functions include:
    - **ReLU (Rectified Linear Unit):** Default for hidden layers in MLPRegressor.
    - **Logistic (Sigmoid):** For non-linear transformations.
    - **Tanh:** Scales outputs between -1 and 1.

- **Optimization Algorithms:**
  - The model uses solvers like **Adam** or **SGD (Stochastic Gradient Descent)** to minimize the loss function.

---

**How It Works:**

1. **Forward Pass:**
   - Input features are passed through the network, with each neuron applying a weighted sum of inputs followed by an activation function.

2. **Loss Calculation:**
   - The model calculates the error between predicted and actual target values using a loss function like Mean Squared Error (MSE).

3. **Backpropagation:**
   - Gradients of the loss function with respect to weights are computed and propagated backward to update the model parameters.

4. **Iteration:**
   - The process is repeated for a fixed number of epochs or until convergence, enabling the model to learn the relationships in the data.

---

**Advantages:**

- **Captures Non-Linear Relationships:** The use of hidden layers and activation functions allows MLPRegressor to learn complex, non-linear patterns.

- **Flexible Architecture:** Supports multiple layers and neurons, enabling customization for different data complexities.
- **Wide Applicability:** Works well on structured data for tasks like price prediction, forecasting, and more.

---

**When to Use:**

- **Non-Linear Data:** When the relationship between features and the target variable is complex and non-linear.

- **Structured Data:** Ideal for tabular datasets where relationships between features are intricate.
- **Large Datasets:** Performs well with sufficient data for training, as it requires large amounts of information to avoid overfitting.

---

**Summary:**

MLPRegressor is a robust regression algorithm that leverages the power of neural networks to model complex, non-linear relationships. It uses backpropagation and gradient-based optimization to learn from data, making it suitable for a wide range of regression tasks. 2. **param_grid_mlp:**

- **Purpose:**
  - Defines a dictionary of hyperparameters to explore during hyperparameter tuning.
  - Each key corresponds to a hyperparameter of the MLPRegressor model, and the associated values represent the options to be tested.

---

**Parameters of MLPRegressor:**

- **hidden_layer_sizes:**
  - Specifies the architecture of the neural network, defined as the number of neurons in each hidden layer.
  - Options in the grid include:
    - (50,): One hidden layer with 50 neurons.
    - (100,): One hidden layer with 100 neurons.
    - (100, 50): Two hidden layers with 100 and 50 neurons, respectively.
    - (150, 100, 50): Three hidden layers with 150, 100, and 50 neurons.

- **activation:**
  - Defines the activation function for the hidden layers:
    - 'relu': Rectified Linear Unit (default), introduces non-linearity while preventing vanishing gradients.
    - 'tanh': Hyperbolic tangent, scales output between -1 and 1.
    - 'logistic': Sigmoid function, scales output between 0 and 1.

- **solver:**
  - The optimization algorithm for updating weights:
    - 'adam': A robust gradient-based optimizer that works well in practice.
    - 'sgd': Stochastic Gradient Descent, which updates weights incrementally for each data point.

- **alpha:**
  - L2 regularization term to prevent overfitting by penalizing large weights.
  - Options in the grid: [0.0001, 0.001, 0.01, 0.1]

- **learning_rate:**
  - The schedule for adjusting the learning rate during training:
    - 'constant': Keeps the learning rate fixed throughout training.
    - 'adaptive': Adjusts the learning rate dynamically based on training progress.

- **learning_rate_init:**
  - The initial learning rate for weight updates.
  - Options in the grid: [0.001, 0.01, 0.1]

- **max_iter:**
  - The maximum number of iterations allowed for training.
  - Options in the grid: [200, 400, 600]

---

- The **MLPRegressor** instance (mlp) is initialized with default parameters, ready to be tuned and trained.

- Each hyperparameter influences the neural network's architecture, optimization, and learning process, enabling it to capture complex patterns while avoiding overfitting.

**1. RandomizedSearchCV Initialization:**

- **Purpose:**
  - To perform a randomized search over a specified grid of hyperparameters for the **MLPRegressor (Multi-Layer Perceptron Regressor)**.

  - The RandomizedSearchCV object evaluates a fixed number of randomly selected combinations from the parameter grid (param_grid_mlp) and selects the one with the best performance.

---

**Parameters Explained:**

- **estimator=mlp:**
  - The **MLPRegressor** is used as the base model to be tuned.

- **param_distributions=param_grid_mlp:**
  - Passes the hyperparameter grid (param_grid_mlp) containing possible values for:
    - hidden_layer_sizes: Neural network architectures with different numbers of neurons and layers.

    - activation: Activation functions for the hidden layers, such as ReLU, tanh, and logistic.
    - solver: Optimization algorithms for weight updates (e.g., Adam or SGD).
    - alpha: L2 regularization parameter to prevent overfitting.
    - learning_rate: Learning rate schedule, either constant or adaptive.
    - learning_rate_init: Initial learning rate for the optimization algorithm.
    - max_iter: Maximum number of iterations allowed for training.

  - This grid defines the hyperparameter combinations to be randomly sampled during the search.

- **cv=3:**
  - Specifies 3-fold cross-validation:
    - The dataset is split into 3 equal parts.
    - Each hyperparameter combination is evaluated 3 times using different train-test splits.

- **scoring='neg_mean_squared_error':**
  - Specifies that the evaluation metric is the negative mean squared error (MSE), which the algorithm aims to minimize.

- **verbose=3:**
  - Provides detailed output during the search process, displaying progress for each hyperparameter combination.

- **return_train_score=True:**
  - Ensures that training scores are also recorded alongside validation scores for each hyperparameter combination.

---

**2. rand_mlp.fit(x_train_scaled, y_train):**

- **Purpose:**
  - Fits the RandomizedSearchCV object to the scaled training data (x_train_scaled) and target variable (y_train).

- **What Happens During fit():**
  - The model trains and evaluates a fixed number of random hyperparameter combinations from the grid (param_grid_mlp).

  - Cross-validation ensures the model is validated on different splits of the training data.
  - The best combination of hyperparameters is stored internally.

---

**How It Works:**

1. **Random Sampling:**
   - RandomizedSearchCV randomly selects a fixed number of hyperparameter combinations (instead of exhaustively testing all combinations as in GridSearchCV).

2. **For Each Combination:**
   - The model is trained using the selected hyperparameters.
   - Performance is evaluated on the validation set using cross-validation.
   - Training scores are also recorded (because return_train_score=True).

3. **Best Hyperparameter Selection:**
   - The hyperparameter combination that gives the lowest negative mean squared error is stored in rand_mlp.best_params_ .

---

**Key Outputs:**

After the fitting process:

- **rand_mlp.best_params_:** Returns the best combination of hyperparameters.

- **rand_mlp.best_score_:** The best cross-validated score achieved during the search.
- **rand_mlp.cv_results_:** Detailed results for all evaluated combinations, including training and validation scores.

---

**Summary:**

- RandomizedSearchCV efficiently explores a random subset of hyperparameter combinations in param_grid_mlp for the **MLPRegressor** model.

- The fit() method trains and evaluates the model using 3-fold cross-validation.
- The best hyperparameters are identified based on the lowest negative mean squared error, optimizing the model’s performance while saving computational resources compared to exhaustive searches.


```python
rand_mlp = RandomizedSearchCV(estimator=mlp, param_distributions=param_grid_mlp, cv=3, scoring='neg_mean_squared_error', verbose=3, return_train_score=True)
rand_mlp.fit(x_train_scaled, y_train)
```

These metrics are calculated by comparing the actual target values (y_test) with the predicted values (y_pred_test_mlp), and the results are printed for interpretation.

---

1.**mean_squared_error(y_test, y_pred_test_mlp)**

- Purpose:

 - Calculates the Mean Squared Error (MSE) between the actual values (y_test) and the predicted values (y_pred_test_mlp).

- Interpretation:

 - MSE gives the average squared difference between the actual and predicted values. A lower MSE indicates that the model’s predictions are close to the true values.

 - Since the errors are squared, large errors have a bigger impact on the result.

---

2.**mean_absolute_error(y_test, y_pred_test_mlp)**

- Purpose:

 - Calculates the Mean Absolute Error (MAE) between the actual values (y_test) and the predicted values (y_pred_test_mlp).

- Interpretation:

 - MAE measures the average magnitude of the errors in the predictions without considering their direction. A lower MAE indicates better prediction accuracy.

 - Unlike MSE, MAE gives equal weight to all errors, which can be useful when you want to avoid large errors disproportionately affecting the result.

---

3.**r2_score(y_pred_test_mlp, y_test)**

- Purpose:

 - Computes the R-squared (R²) score, which indicates how well the model explains the variance in the target variable.

- Interpretation:

 - R-squared shows the proportion of variance in the target variable that is explained by the model.

 - An R² close to 1 means the model explains most of the variance in the data, while an R² close to 0 indicates that the model doesn’t explain much of the variance.

 - If the R² is negative, it indicates that the model performs worse than simply predicting the mean of the target variable.

---

**Purpose of the Code:**

- These print statements display the computed values for each of the evaluation metrics:

 - MSE: Gives a sense of the average squared error.

 - MAE: Provides the average absolute error, showing the typical magnitude of the model's mistakes.

 - R²: Indicates how well the model fits the data, with higher values showing better model performance.

---

**Summary:**

The code prints the evaluation metrics for the model:

- MSE: Measures the squared error.

- MAE: Measures the absolute error.

- R²: Measures the goodness of fit.
